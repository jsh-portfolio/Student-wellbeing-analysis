```python
import pandas as pd
df = pd.read_csv('../data/student_wellbeing.csv')
df.head()
```

### Cleaning column names

Column names were standardised to make the dataset easier to query and analyse. While also assuring that naming conventions remained consistant across SQL, Python, and Power BI workflows.

```python
df.columns = (
    df.columns
    .str.lower()
    .str.strip()
    .str.replace(' ', '_')
    .str.replace('?', '')
    .str.replace('/', '_')
)
```

```python
df.columns
```

```python
df = df.rename(columns={
    "choose_your_gender": "gender",
    "what_is_your_course": "course",
    "your_current_year_of_study": "year_of_study",
    "what_is_your_cgpa": "cgpa",
    "do_you_have_depression": "depression",
    "do_you_have_anxiety": "anxiety",
    "do_you_have_panic_attack": "panic_attack",
    "did_you_seek_any_specialist_for_a_treatment": "treatment"
})
```

### Selecting relevant variables

Only variables relevant to the project questions were retained. This keeps the analysis focused on demographic, academic, and wellbeing-related patterns rather than attempting to analyse every available column.

```python
df = df[[
    "gender",
    "age",
    "course",
    "year_of_study",
    "cgpa",
    "depression",
    "anxiety",
    "panic_attack",
    "treatment"
]]
```
