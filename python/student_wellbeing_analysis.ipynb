```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('../data/student_wellbeing_cleaned.csv')

df.head()
df.info()
df.describe()
df.isnull().sum()
```

```python
plt.figure(figsize=(6,4))

df['depression'].value_counts().plot(kind='bar')

plt.title('Breakdown of Depression')
plt.xlabel('Depression')
plt.ylabel('Number of Students')

plt.tight_layout()
plt.show()
```

### Analysis Depression

The chart shows that more students reported not experiencing depression than those who did. However, a noticeable proportion of students still reported depression, suggesting that mental health concerns are present within the student sample. This could indicate a need to further explore factors such as anxiety, panic attacks, or treatment support.

```python
df["anxiety"].value_counts().plot(kind="bar")
plt.title("Anxiety Breakdown")
plt.xlabel("Anxiety")
plt.ylabel("Number of Students")
plt.tight_layout()
plt.show()
```

### Analysis Anxiety

The chart shows that more students reported not experiencing anxiety than those who did. However, a considerable number of students still reported anxiety symptoms, suggesting that mental health challenges are present within the dataset.

This may indicate that academic or personal pressures are affecting student wellbeing and could be explored further alongside other variables such as depression and panic attacks.

```python
df["panic_attack"].value_counts().plot(kind="bar")
plt.title("Panic Attack Breakdown")
plt.xlabel("Panic Attack")
plt.ylabel("Number of Students")
plt.tight_layout()
plt.show()
```

### Analysis Panic Attack

The chart shows that most students reported not experiencing panic attacks. However, a noticeable number of students did report experiencing panic attacks, which suggests that mental health difficulties are still affecting a significant portion of the student sample.

This may indicate that factors such as academic stress, anxiety, or overall wellbeing could be contributing to increased levels of psychological pressure among some students.

```python
df["treatment"].value_counts().plot(kind="bar")
plt.title("Treatment Seeking Breakdown")
plt.xlabel("Treatment")
plt.ylabel("Number of Students")
plt.tight_layout()
plt.show()
```

### Analysis Treatment seeking behaviours

The chart shows that only a small number of students reported seeking treatment, while the majority stated that they had not. This could suggest that many students experiencing mental health difficulties may not be accessing professional support or treatment services.

A possible explanation could be lack of awareness, stigma surrounding mental health, or students not recognising the severity of their symptoms. This highlights the importance of mental health support and accessibility within educational environments.

```python
df["any_concern"] = (
    (df["depression"] == "Yes") |
    (df["anxiety"] == "Yes") |
    (df["panic_attack"] == "Yes")
)

df["any_concern"].value_counts()
```

```python
df["number_of_concerns"] = (
    (df["depression"] == "Yes").astype(int) +
    (df["anxiety"] == "Yes").astype(int) +
    (df["panic_attack"] == "Yes").astype(int)
)

plt.figure(figsize=(6,4))

df["number_of_concerns"] \
    .value_counts() \
    .sort_index() \
    .plot(kind="bar")

plt.title("Number of Mental Health Concerns Reported")
plt.xlabel("Number of Reported Concerns")
plt.ylabel("Number of Students")

plt.tight_layout()
plt.show()
```

### Analysis Comorbidity overlap within data sample

The chart shows that many students reported at least one mental health concern, while a smaller number reported experiencing multiple concerns at the same time. A noticeable group of students reported two or more wellbeing difficulties simultaneously.

This may suggest that mental health challenges such as depression, anxiety, and panic attacks can overlap for some students, potentially increasing the overall impact on student wellbeing and daily functioning.
